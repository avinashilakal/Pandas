#### Groupby on the data

Doing groupby on Pclass based on Survival status.
Pclass contains 3 different values 1,2 3 and Survived contains two different values 0,1.
From below output we are geeting How many percent people survived in each class.

```python
df[['Pclass','Survived']].groupby(['Pclass'],as_index=False).mean()
```

```python
df.groupby('Pclass').size()
```
